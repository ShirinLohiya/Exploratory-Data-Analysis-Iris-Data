# MEAN

```python
iris_setosa = iris[iris["species"] == "setosa"]
mean_petal_length = np.mean(iris_setosa["petal_length"])
print("Mean petal length of Setosa:", mean_petal_length)

# Append 50 to petal_length and compute new mean
new_petal_lengths = np.append(iris_setosa["petal_length"], 50)
new_mean_petal_length = np.mean(new_petal_lengths)
print("New Mean petal length after adding 50:", new_mean_petal_length)
```
